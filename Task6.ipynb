```python
class Sack:
    def __init__(self, contents, weight):
        self.contents = contents
        self.weight = weight

def check_sack(sack):
    if sack.contents not in ['C', 'G', 'S']:
        return "Invalid contents. Accepted values: C, G, S."

    if sack.contents == 'C' and not (24.9 < sack.weight < 25.1):
        return "Invalid weight for cement sack."

    if sack.contents in ['G', 'S'] and not (49.9 < sack.weight < 50.1):
        return "Invalid weight for gravel or sand sack."

    return f"Sack accepted - Contents: {sack.contents}, Weight: {sack.weight} kg."

def check_order(order):
    total_weight = 0
    sacks_rejected = 0

    for contents, quantity in order.items():
        for _ in range(quantity):
            weight = float(input(f"Enter the weight for a {contents} sack: "))
            sack = Sack(contents, weight)
            result = check_sack(sack)

            if "accepted" in result:
                total_weight += sack.weight
            else:
                sacks_rejected += 1
                print(result)

    return total_weight, sacks_rejected

def calculate_price(total_weight, order):
    prices = {'C': 3, 'G': 2, 'S': 2}
    special_pack_price = 10

    regular_price = sum(order[contents] * prices[contents] for contents in order)
    special_packs = min(order.get('C', 0) // 1, order.get('S', 0) // 2, order.get('G', 0) // 2)

    discount_price = regular_price - special_packs * special_pack_price
    amount_saved = regular_price - discount_price

    return regular_price, discount_price, amount_saved

```

```python
# TASK 1
contents = input("Enter the contents for a sack (C, G, S): ")
weight = float(input("Enter the weight for the sack: "))
sack = Sack(contents, weight)
result = check_sack(sack)
print(result)

```

```python

# TASK 2
order = {'C': 2, 'G': 3, 'S': 1}  # Example order
total_weight, sacks_rejected = check_order(order)
print(f"Total weight of the order: {total_weight} kg.")
print(f"Number of sacks rejected from the order: {sacks_rejected}")

```

```python
# TASK 3
regular_price, discount_price, amount_saved = calculate_price(total_weight, order)
print(f"Regular price for the order: ${regular_price}")
if discount_price > 0:
    print(f"Discounted price for the order: ${discount_price} (You saved ${amount_saved})")
```
